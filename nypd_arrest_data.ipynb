```python
import pandas as pd
```

```python
#import datasets
df1 = pd.read_csv('NYPD_Arrest_Data_2023.csv')
df2 = pd.read_csv('NYPD_Arrests_Data__Historic.csv')
```

```python
#change column name in df1 to match df2
df1.rename(columns={'New Georeferenced Column': 'Lon_Lat'}, inplace=True)
```

```python
#concat datasets together
crime_dataset = pd.concat([df1, df2], ignore_index=True)
```

```python
#remove duplicates based on all columns
crime_dataset.drop_duplicates()
```

```python
#drop rows with blanks
crime_dataset.dropna()
```

```python
#drop unneeded columns
crime_dataset.drop(['PD_CD', 'KY_CD', 'JURISDICTION_CODE'], axis=1, inplace=True)
```

```python
#rename column names to make it more intuitive
crime_dataset.rename(columns={
    'LAW_CAT_CD' : 'LEVEL_OF_OFFENSE',
    'ARREST_BORO' : 'ARREST_BOROUGH',
    'PD_DESC' : 'POLICE_DESC',
    'OFNS_DESC' : 'OFFENSE_DESC'
}, inplace=True)
```

```python
#rename Values to make it more intuitive
df = crime_dataset.replace({
    'LEVEL OF OFFENSE': {
        'F': 'Felony',
        'M' : 'Misdemeanor',
        'V' : 'Violation'
    }, 
    'ARREST_BOROUGH': {
        'B': 'Bronx',
        'S': 'Staten Island',
        'K': 'Brooklyn',
        'M': 'Manhattan',
        'Q': 'Queens',
    },
    'PERP_SEX' : {
        'M' : 'Male',
        'F' : 'Female'
    }
})
```

```python
#clean AGE_GROUP column
age_brackets = ['25-44', '<18', '45-64', '65+']

df = df[df['AGE_GROUP'].isin(age_brackets)]

df['AGE_GROUP'].unique()
```

```python
#export to csv
df.to_csv('nypd_crime_2006_to_2023.csv', index=False)
```
